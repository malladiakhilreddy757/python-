# Slicing

```python
#[start:stop:step]
#here start = starting position 
#stop is the ending position 
# step = step size or increment size (defalt value is 1)
```

```python
# ex 1
a = 'hello world'# to grabe the hello word we have done slicing
a[0:5:1]
```

```python
a[6:]

```

```python
a[-5:]
```

```python
a[::2]#to get the alternate numbers 
```

```python
a[::-1] # to get the reverse string
```
